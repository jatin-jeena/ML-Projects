```python
product = ['Ipod', 'Mobiles', 'Keyboard', 'Mouse', 'Notebook', 'Lan', 'Hard Drive' ,'PenDrive', 'Earphones', 'Speakers']
```

```python
product_price = {'Ipod':2000, 'Mobiles':15000, 'Keyboard':1000, 'Mouse':900, 'Notebook':75000, 'Lan':1200, 'Hard Drive':5000 ,'PenDrive':2500, 'Earphones':2599, 'Speakers':5500}
```

```python
import random

sales['Final_price'] = 0
```

```python
for i, j in sales.iterrows():
    x = random.choice(product)
    sales.loc[i, 'ProductName'] = x
    price = product_price[x]
    sales.loc[i, 'Price'] = price
    city = j['Region']
    ship_price = shipping_price[city]
    final_price = price + (price*ship_price)/100
    sales.loc[i, 'Final_price'] = final_price
```
